```python
import numpy as np
import matplotlib.pylab as plt

%matplotlib notebook
```

```python
# R = np.arange(1,5,1) # Ohms????
V = np.arange(0,12,0.1)
I = 150 - V**2
```

```python
plt.figure()
plt.plot(I,V)
```

```python
P = I*V
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(V,P)
plt.xlabel("Voltage (V)")
plt.ylabel("Power (Watts)")

plt.subplot(1,2,2)
plt.plot(I,P)
plt.xlabel("Current (A)")
plt.ylabel("Power (Watts)")
```
