```python
import numpy as np
import matplotlib.pylab as plt

%matplotlib notebook

import seaborn as sns
```

```python
# Resistance values
R = np.array([188.0,126.3,92.8,75.0,21.2,16.2,5.05,2.65]);

# PANEL ONE
V1_away = np.array([4.88,5.15,4.10,4.08,2.02,0.67,0.29,0.10]);
I1_away = 0.001*np.array([0.26,0.38,0.56,0.66,0.59,0.58,0.57,0.59]);
P1_away = V1_away * I1_away

V1_toward = np.array([5.81,5.93,5.78,5.74,5.46,5.24,2.98,1.48]);
I1_toward = 0.001*np.array([0.30,0.45,0.62,0.76,2.54,3.27,5.80,5.71]);
P1_toward = V1_toward * I1_toward

# PANEL TWO
V2_away = np.array([4.43,4.12,3.28,2.75,0.77,0.66,0.19,0.18]);
I2_away = 0.001*np.array([0.23,0.31,0.34,0.36,0.40,0.37,0.54,0.97]);
P2_away = V2_away * I2_away

V2_toward = np.array([5.76,5.71,5.63,5.63,5.35,5.25,2.90,1.50]);
I2_toward = 0.001*np.array([0.30,0.44,0.60,0.74,2.51,3.24,5.79,5.64]);
P2_toward = V2_toward * I2_toward
```

```python
plt.figure(figsize=(12,12))
plt.subplot(3,2,1)
plt.plot(R,V1_away,"rs",markersize=10,label="Panel 1 -away")
plt.plot(R,V1_toward,"r*",markersize=20,label="Panel 1 -toward")
plt.plot(R,V2_away,"bs",markersize=10,label="Panel 2 -away")
plt.plot(R,V2_toward,"b*",markersize=20,label="Panel 2 -toward")
plt.xlabel(r"Resistance ($\Omega$)",fontsize=18)
plt.ylabel(r"Voltage (V)",fontsize=18)
plt.ylim(0,8)
plt.legend(loc='lower right')


plt.subplot(3,2,2)
plt.plot(R,I1_away,"rs",markersize=10,label="Panel 1 -away")
plt.plot(R,I1_toward,"r*",markersize=20,label="Panel 1 -toward")
plt.plot(R,I2_away,"bs",markersize=10,label="Panel 2 -away")
plt.plot(R,I2_toward,"b*",markersize=20,label="Panel 2 -toward")
plt.xlabel(r"Resistance ($\Omega$)",fontsize=18)
plt.ylabel(r"Current (A)",fontsize=18)
#plt.ylim(0,9)
plt.legend(loc='upper right')


plt.subplot(3,2,3)
plt.plot(V1_away,I1_away,"rs",markersize=10,label="Panel 1 -away")
plt.plot(V1_toward,I1_toward,"r*",markersize=20,label="Panel 1 -toward")
plt.plot(V2_away,I2_away,"bs",markersize=10,label="Panel 2 -away")
plt.plot(V2_toward,I2_toward,"b*",markersize=20,label="Panel 2 -toward")
plt.xlabel(r"Voltage (V)",fontsize=18)
plt.ylabel(r"Current (A)",fontsize=18)
plt.ylim(0,0.009)
plt.legend(loc='upper right')


plt.subplot(3,2,4)
plt.plot(R,P1_away,"rs",markersize=10,label="Panel 1 -away")
plt.plot(R,P1_toward,"r*",markersize=20,label="Panel 1 -toward")
plt.plot(R,P2_away,"bs",markersize=10,label="Panel 2 -away")
plt.plot(R,P2_toward,"b*",markersize=20,label="Panel 2 -toward")
plt.xlabel(r"Resistance($\Omega$)",fontsize=18)
plt.ylabel(r"Power (W)",fontsize=18)
#plt.ylim(0,9)
plt.legend(loc='upper right')

plt.subplot(3,2,5)
plt.plot(V1_away,P1_away,"rs",markersize=10,label="Panel 1 -away")
plt.plot(V1_toward,P1_toward,"r*",markersize=20,label="Panel 1 -toward")
plt.plot(V2_away,P2_away,"bs",markersize=10,label="Panel 2 -away")
plt.plot(V2_toward,P2_toward,"b*",markersize=20,label="Panel 2 -toward")
plt.xlabel(r"Voltage(V)",fontsize=18)
plt.ylabel(r"Power (W)",fontsize=18)
#plt.ylim(0,9)
plt.legend(loc='upper left')

plt.subplot(3,2,6)
plt.plot(I1_away,P1_away,"rs",markersize=10,label="Panel 1 -away")
plt.plot(I1_toward,P1_toward,"r*",markersize=20,label="Panel 1 -toward")
plt.plot(I2_away,P2_away,"bs",markersize=10,label="Panel 2 -away")
plt.plot(I2_toward,P2_toward,"b*",markersize=20,label="Panel 2 -toward")
plt.xlabel(r"Current(A)",fontsize=18)
plt.ylabel(r"Power (W)",fontsize=18)
#plt.ylim(0,9)
plt.legend(loc='upper left')


plt.tight_layout()
```
